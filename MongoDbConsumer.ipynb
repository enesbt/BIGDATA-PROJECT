```python

from pymongo import MongoClient
#topice eklenen verileri test
from confluent_kafka import Consumer, KafkaException
import sys
# Kafka consumer ayarları

client = MongoClient('localhost', 27017)
db = client['KafkaConsume']
collection = db['MobilePriceTestData']



# Kafka broker'ının adresi ve portu
bootstrap_servers = 'localhost:9092'

# Kafka konusu
topic = 'mobilPriceTopic'

# Consumer oluşturma
conf = {'bootstrap.servers': bootstrap_servers, 'group.id': 'mygroup', 'auto.offset.reset': 'earliest'}
consumer = Consumer(conf)

# Kafka konusuna abone olma
consumer.subscribe([topic])

try:
    while True:
        # Mesajları tüketme
        msg = consumer.poll(timeout=1.0)

        if msg is None:
            continue
        if msg.error():
            if msg.error().code() == KafkaException._PARTITION_EOF:
                continue
            else:
                print(msg.error())
                break

        try:
           collection.insert_one({"message": msg.value().decode('utf-8')})
           print("eklendi")
        except Exception as e:
            print("Hata:", e)

except KeyboardInterrupt:
    sys.exit(0)

finally:
    # Consumer'ı kapatma
    consumer.close()

```
